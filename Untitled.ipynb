```python
# # This script takes a list of urls and downloads the file
# from each url to a folder, after naming the file


import os
import re
import csv
import urllib
import re
  
# create folder for saving signatures
folders = os.listdir('./')
savefolder = 'AG_182_FULL_sigs'
if savefolder not in folders:
    os.mkdir(savefolder)



urls=[]
data = csv.reader(open('AG_FULL_DEGsignatures.csv','rU'), delimiter=',')    
for line in data:
    urls.append(line[0])
    for url in urls:
        htmlfile =  urllib.urlopen(url)
        htmltext =  htmlfile.readlines() 
        print(url)
        fname=url.strip('http://amp.pharm.mssm.edu/g2e/gene_list/0/')
        print (fname)
        # GEN3VA base urls:
        # complete list http://amp.pharm.mssm.edu/g2e/gene_list/0/
        # Up list http://amp.pharm.mssm.edu/g2e/gene_list/1/
        #Down list http://amp.pharm.mssm.edu/g2e/gene_list/-1/

    # specify prefix of output files
        savefileprefix = 'full_'
        savefile = savefolder + '/' + savefileprefix + fname + '.txt'
        with open(savefile, mode='w') as fw:
            for line in htmltext:
                fw.write(line)
```
